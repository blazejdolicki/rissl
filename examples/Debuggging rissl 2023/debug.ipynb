```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test_probs = pd.read_csv("test_probs.csv")[["label", "prob0", "prob1"]]
test_probs.head()
```

```python
sns.kdeplot(data=test_probs, x="prob1", hue="label", common_norm=True, fill=True)
plt.title("Distribution of BreaKHis test set predicted probabilities")
plt.show()
```

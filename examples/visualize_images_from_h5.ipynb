```python
import numpy as np  
import matplotlib.pyplot as plt  
import h5py
import os

# load h5py images and labels
DATA_PATH = "data/pcam/"
images = h5py.File(os.path.join(DATA_PATH,'camelyonpatch_level_2_split_test_x.h5'),'r')
images = images["x"]

labels = h5py.File(os.path.join(DATA_PATH,'camelyonpatch_level_2_split_test_y.h5'),'r')
labels = labels["y"]

label_names = {0:"healthy", 1:"tumour"}
```

```python
num_images = 12 # must be dividable by 4

plt.subplots(num_images//4,4,figsize = (20,20))
plt.suptitle(f"First {num_images} images with corresponding labels", y=0.90, fontsize=18)
for i in range(num_images):
    img_np = images[i]
    plt.subplot(num_images//4,4,1+i)
    plt.title(f"label: {label_names[labels[i].item()]}")
    plt.imshow(img_np)
plt.show()
```
